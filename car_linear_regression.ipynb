```python
mat = cars_copy.to_numpy()
mat.shape
```

```python
X = np.zeros([mat.shape[0],mat.shape[1]])
X[:,0] = np.ones([mat.shape[0]])
X[:,1:mat.shape[1]] = mat[:,0:mat.shape[1]-1]
X
```

```python
Y = mat[:,-1]
Y = Y.reshape([205,1])
m = Y.shape[0]
```

```python
def feature_normalize(X):
    X_norm = np.zeros([X.shape[0],X.shape[1]])
    X_norm[:,0:16] = X[:,0:16]
    X_mean = np.zeros([X.shape[0],X.shape[1]])
    X_std = np.ones([X.shape[0],X.shape[1]])
    for i in range(1,16):
        X_mean[:,i] = np.mean(X_norm[:,i])
        X_std[:,i] = np.std(X_norm[:,i])
    X_norm = (X_norm-X_mean)/X_std
    X_norm[:,17:] = X[:,17:]
    return X_norm
```

```python
X = feature_normalize(X)
X
```

```python
theta = np.zeros([X.shape[1],1])
iterations = 2500
alpha = 0.001
    
```

```python
def cost_function(X,Y,theta):
    return (1/(2*m))*(np.sum((np.matmul(X,theta)-Y)**2))
```

```python
def gradient_descent(X,Y,theta,alpha,iterations):
    J_hist = np.zeros([iterations,1])
    for i in range(iterations):
        theta = theta-(alpha/m)*(np.matmul(np.transpose(X),np.matmul(X,theta)-Y))
        J_hist[i] = cost_function(X,Y,theta)
    return J_hist,theta
```

```python
J,theta = gradient_descent(X,Y,theta,alpha,iterations)
```

```python
plt.plot(J)
```
